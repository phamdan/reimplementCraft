```
import scipy.io
from scipy.io import savemat
import numpy as np
import os

gt = scipy.io.loadmat('./pretrain/data/CRAFT-pytorch/Syndata/gt.mat')
charbox = gt['charBB'][0]
image = gt['imnames'][0]
imgtxt = gt['txt'][0]
img_name=[]
path="./pretrain/data/CRAFT-pytorch/Syndata"
folder=os.listdir("./pretrain/data/CRAFT-pytorch/Syndata")
folder.remove('gt.mat')
for value in folder:
  sub_folder=os.listdir(os.path.join(path,value))
  for i in range(len(sub_folder)):
    sub_folder[i]=value+"/"+sub_folder[i]
  img_name+=sub_folder
index = np.flatnonzero(np.isin(image, img_name))
number_image=np.arange(len(image))
index_image_remove=np.delete(number_image,index)
image=np.delete(image,index_image_remove)
charbox=np.delete(charbox,index_image_remove)
imgtxt=np.delete(imgtxt,index_image_remove)
```
